```python
import matplotlib.pyplot as plt

# Data for Majority Voting Approach
labels_majority = ['Negative', 'Neutral', 'Positive']
sizes_majority = [5194, 2481, 229]
total_majority = sum(sizes_majority)
percentages_majority = [(size / total_majority) * 100 for size in sizes_majority]

# Data for Weighted Voting Approach
labels_weighted = ['Negative', 'Neutral', 'Positive']
sizes_weighted = [62, 6992, 850]
total_weighted = sum(sizes_weighted)
percentages_weighted = [(size / total_weighted) * 100 for size in sizes_weighted]

# Data for Gradient Boosting Approach
labels_gb = ['Negative', 'Neutral', 'Positive']
sizes_gb = [6278, 1562, 64]
total_gb = sum(sizes_gb)
percentages_gb = [(size / total_gb) * 100 for size in sizes_gb]

# Create bar chart for Majority Voting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_majority, percentages_majority)
plt.title('Majority Voting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.savefig('majority_voting_chart.png')
plt.show()

# Create bar chart for Weighted Voting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_weighted, percentages_weighted)
plt.title('Weighted Voting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.savefig('weighted_voting_chart.png')
plt.show()

# Create bar chart for Gradient Boosting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_gb, percentages_gb)
plt.title('Gradient Boosting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.savefig('gradient_boosting_chart.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data for Majority Voting Approach
labels_majority = ['Negative', 'Neutral', 'Positive']
sizes_majority = [5194, 2481, 229]
total_majority = sum(sizes_majority)
percentages_majority = [(size / total_majority) * 100 for size in sizes_majority]

# Data for Weighted Voting Approach
labels_weighted = ['Negative', 'Neutral', 'Positive']
sizes_weighted = [62, 6992, 850]
total_weighted = sum(sizes_weighted)
percentages_weighted = [(size / total_weighted) * 100 for size in sizes_weighted]

# Data for Gradient Boosting Approach
labels_gb = ['Negative', 'Neutral', 'Positive']
sizes_gb = [6278, 1562, 64]
total_gb = sum(sizes_gb)
percentages_gb = [(size / total_gb) * 100 for size in sizes_gb]

# Custom color palette for the bar charts
colors = sns.color_palette('pastel')

# Set seaborn style (optional but adds some additional beauty)
sns.set(style='whitegrid', font_scale=1.2)

# Create bar chart for Majority Voting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_majority, percentages_majority, color=colors)
plt.title('Majority Voting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.savefig('majority_voting_chart.png')
plt.show()

# Create bar chart for Weighted Voting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_weighted, percentages_weighted, color=colors)
plt.title('Weighted Voting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('weighted_voting_chart.png')
plt.show()

# Create bar chart for Gradient Boosting Approach
plt.figure(figsize=(8, 6))
plt.bar(labels_gb, percentages_gb, color=colors)
plt.title('Gradient Boosting Approach - Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('gradient_boosting_chart.png')
plt.show()
```
